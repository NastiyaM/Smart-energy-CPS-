```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
results  = pd.read_csv("results_GLOP_S.csv",sep = ',')
results
```

```python
money_spent = []
money_no_batt = []
score = []
for i in np.unique(results['site_id']):
    money_spent.append(sum(results[results.site_id == i]['money_spent']))
    money_no_batt.append(sum(results[results.site_id == i]['money_no_batt']))
    score.append([i,np.mean(results[results.site_id == i]['score'])])
```

```python
score
```

```python
index = np.arange(11)
values1 = money_spent
values2 = money_no_batt
bw = 0.4
plt.figure(figsize=(12, 7))
plt.bar(index, values1, bw,label='Затрат с батареей')
plt.bar(index+bw, values2, bw,label='Затраты без батареи')
plt.legend()
plt.show()
```

```python
x = list(i for i in range(len(results[results.site_id == 12]['money_no_batt'])))
plt.figure(figsize=(12, 7))
plt.plot(x,results[results.site_id == 12]['money_spent'], alpha=0.7, label="Затрат с батареей", lw=5, mec='b', mew=2, ms=10)
plt.plot(x, results[results.site_id == 12]['money_no_batt'], label="Затраты без батареи", mec='r', lw=2, mew=2, ms=12)
plt.legend()
plt.grid(True)
```
